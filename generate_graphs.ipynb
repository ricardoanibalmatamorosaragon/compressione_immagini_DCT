```python
#i commenti vanno tolti per ottenere i file contenenti i parametri riusltanti da ogni esecuzione.
from dct2 import dct2, my_dct2
import numpy as np
import time 
import plotly.plotly as py
import plotly
import matplotlib.pyplot as plt

def gen_matrix(start,end,rows,cols):
    a=np.random.randint(start,end, size=(rows,cols))
    return a

def main():
    np.random.seed(33)
    fig = plt.figure()
    ax = plt.subplot(111)
    y_time_mydct=[]
    y_time_dct=[]
    x_size=[]
    #f = open("time_dct2.txt",'w')
    #f = open("time_dct2_step1.txt",'w')
    for i in range(2,101):
        #matrix=gen_matrix(1,255,100*i,100*i)
        matrix=gen_matrix(1,255,i,i)
        start_dct2=time.clock()
        dct2(matrix)
        end_dct2=time.clock() - start_dct2
        start_my_dct2=time.clock()
        my_dct2(matrix)
        end_my_dct2=time.clock() - start_my_dct2


       
        y_time_mydct.append(end_my_dct2)
        y_time_dct.append(end_dct2)
        x_size.append(i)
    ax.plot(x_size,y_time_mydct, label ='MyDCT2')
    ax.plot(x_size,y_time_dct, label ='DCT2')
    # Shrink current axis by 20%
    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width * 3, box.height* 3])

    # Put a legend to the right of the current axis
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.title('Tempo computazionale')
    plt.ylabel('Time')
    plt.xlabel('matrix size')
    plt.grid()
    
    plt.yscale('log')
    plt.show()
    '''
    f.write("size matrix "+str(i)+" : "+str(len(matrix))+'\n')
    f.write("time to DCT2 python : "+str(end_dct2)+'\n')
    f.write("time to My_DCT2 python : "+str(end_my_dct2)+'\n')
    f.write('\n')    
    '''


```
